```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('rxns_of_interest.csv')
mols_to_pred_shap = []
for _, row in df.iterrows():
    rxnidx, r1, r2, p1, p2, solvent = row['reaction_index'], row['reactant1_smiles'], row['reactant2_smiles'], row['product1_smiles'], row['product2_smiles'], row['solvent_smiles']

    mols_to_pred_shap.append([r1, solvent, 0.0, 'rxn'+str(rxnidx)+'_r1' ])
    mols_to_pred_shap.append([p1, solvent, 0.0, 'rxn'+str(rxnidx)+'_p1' ])
    
    if not str(r2) == 'nan':
        mols_to_pred_shap.append([r2, solvent, 0.0, 'rxn'+str(rxnidx)+'_r2' ])
    if not str(p2) == 'nan':
        mols_to_pred_shap.append([p2, solvent, 0.0, 'rxn'+str(rxnidx)+'_p2' ])
```

```python
mols_to_pred_shap
```

```python
pd.DataFrame(mols_to_pred_shap, columns = ['can_smiles_solute', 'can_smiles_solvent', 'DGsolv', 'label']).to_csv('molecules_to_predict.csv', index=False)
```
