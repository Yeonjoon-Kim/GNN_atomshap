```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_rxns = pd.read_csv('GreenGroup_solvent_rxnrates_expt_data_collected1.csv')
df_dgsolv = pd.read_csv('molecules_to_predict_results(mjy).csv')
```

```python
reaction_data = {}
```

```python
df_dgsolv_new = pd.DataFrame(columns = df_dgsolv.columns)

for _, row in df_rxns.iterrows():
    r1, r2, p1, p2, solvent, solvent_name = row['reactant1_smiles'], row['reactant2_smiles'], row['product1_smiles'], row['product2_smiles'], row['solvent_smiles'], row['solvent_name']
    reaction_index = row['reaction_index']
    krel_value = row['krel (kliq_to_kref)']

    # r1_dgsolv
    try:
        r1_dgsolv = df_dgsolv[ (df_dgsolv.can_smiles_solute == r1) & (df_dgsolv.can_smiles_solvent == solvent)].predicted.iloc[0]
        pd.concat([df_dgsolv_new, df_dgsolv[ (df_dgsolv.can_smiles_solute == r1) & (df_dgsolv.can_smiles_solvent == solvent)]])
    except:
        r1_dgsolv = 0.0

    #print(r1, solvent, r1_dgsolv)

    # r2_dgsolv
    try:
        r2_dgsolv = df_dgsolv[ (df_dgsolv.can_smiles_solute == r2) & (df_dgsolv.can_smiles_solvent == solvent)].predicted.iloc[0]
    except:
        r2_dgsolv = 0.0
        
    #print(r2, solvent, r2_dgsolv)    

    # p1_dgsolv
    try:
        p1_dgsolv = df_dgsolv[ (df_dgsolv.can_smiles_solute == p1) & (df_dgsolv.can_smiles_solvent == solvent)].predicted.iloc[0]
    except:
        p1_dgsolv = 0.0

    #print(p1, solvent, p1_dgsolv)

    # p2_dgsolv
    try:
        p2_dgsolv = df_dgsolv[ (df_dgsolv.can_smiles_solute == p2) & (df_dgsolv.can_smiles_solvent == solvent)].predicted.iloc[0]
    except:
        p2_dgsolv = 0.0

    
    #print(p2, solvent, p2_dgsolv)

    # Print the total values
    R_dgsolv_total = r1_dgsolv + r2_dgsolv
    P_dgsolv_total = p1_dgsolv + p2_dgsolv
    delta_dgsolv = P_dgsolv_total - R_dgsolv_total

    #print(f"R_dgsolv_total (Reactants): {R_dgsolv_total}")
    #print(f"P_dgsolv_total (Products): {P_dgsolv_total}")
    #print(f"delta_dgsolv (P - R): {delta_dgsolv}")

    krel_value = row['krel (kliq_to_kref)']
    #print(f"krel (kliq_to_kref): {krel_value}")

    # ln(krel) 계산
    try:
        ln_krel = np.log(krel_value)
    except ValueError:
        ln_krel = np.nan

    #print(f"ln(krel (kliq_to_kref)): {ln_krel}")
    #print("-" * 50)

    # reaction_index로 묶어서 데이터 저장
    if reaction_index not in reaction_data:
        reaction_data[reaction_index] = {
            'R_dgsolv_total': [],
            'P_dgsolv_total': [],
            'delta_dgsolv': [],
            'ln_krel': [],
            'solvent_name': []
        }
    reaction_data[reaction_index]['R_dgsolv_total'].append(R_dgsolv_total)
    reaction_data[reaction_index]['P_dgsolv_total'].append(P_dgsolv_total)
    reaction_data[reaction_index]['delta_dgsolv'].append(delta_dgsolv)
    reaction_data[reaction_index]['ln_krel'].append(ln_krel)
    reaction_data[reaction_index]['solvent_name'].append(solvent_name)

'''
from scipy.stats import pearsonr

for reaction_index, data in reaction_data.items():

    df_plot = pd.DataFrame({key: data[key] for key in ['R_dgsolv_total', 'P_dgsolv_total', 'delta_dgsolv', 'ln_krel', 'solvent_name']})

    # 1. R_dgsolv_total vs ln(krel)
    corr_r, _ = pearsonr(df_plot['R_dgsolv_total'], df_plot['ln_krel'])
    plt.figure(figsize=(5, 4))
    sns.scatterplot(x='R_dgsolv_total', y='ln_krel', data=df_plot, s=50)
    for i in range(len(df_plot)):
        plt.text(df_plot['R_dgsolv_total'][i], df_plot['ln_krel'][i], df_plot['solvent_name'][i], fontsize=9)
    # 그래프 안쪽에 상관계수 표시
    plt.text(0.95, 0.05, f"Pearson r = {corr_r:.2f}", 
             fontsize=12, transform=plt.gca().transAxes, ha='right', va='bottom')
    plt.title(f'Reaction {reaction_index}: R_dgsolv_total vs ln(krel)')
    plt.xlabel('R_dgsolv_total (Reactants)')
    plt.ylabel('ln(krel)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 2. P_dgsolv_total vs ln(krel)
    corr_p, _ = pearsonr(df_plot['P_dgsolv_total'], df_plot['ln_krel'])
    plt.figure(figsize=(5, 4))
    sns.scatterplot(x='P_dgsolv_total', y='ln_krel', data=df_plot, s=50)
    for i in range(len(df_plot)):
        plt.text(df_plot['P_dgsolv_total'][i], df_plot['ln_krel'][i], df_plot['solvent_name'][i], fontsize=9)
    # 그래프 안쪽에 상관계수 표시
    plt.text(0.95, 0.05, f"Pearson r = {corr_p:.2f}", 
             fontsize=12, transform=plt.gca().transAxes, ha='right', va='bottom')
    plt.title(f'Reaction {reaction_index}: P_dgsolv_total vs ln(krel)')
    plt.xlabel('P_dgsolv_total (Products)')
    plt.ylabel('ln(krel)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 3. ΔGsolv (P - R) vs ln(krel)
    corr_delta, _ = pearsonr(df_plot['delta_dgsolv'], df_plot['ln_krel'])
    plt.figure(figsize=(5, 4))
    sns.scatterplot(x='delta_dgsolv', y='ln_krel', data=df_plot, s=50)
    for i in range(len(df_plot)):
        plt.text(df_plot['delta_dgsolv'][i], df_plot['ln_krel'][i], df_plot['solvent_name'][i], fontsize=9)
    # 그래프 안쪽에 상관계수 표시
    plt.text(0.95, 0.05, f"Pearson r = {corr_delta:.2f}", 
             fontsize=12, transform=plt.gca().transAxes, ha='right', va='bottom')
    plt.title(f'Reaction {reaction_index}: ΔGsolv (P - R) vs ln(krel)')
    plt.xlabel('ΔGsolv (P - R)')
    plt.ylabel('ln(krel)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
'''
```
