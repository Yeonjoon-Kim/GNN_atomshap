```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('shap_results.csv')
df
```

```python
Shap_contrib = []
pred_from_sum_shap = []
for _, row in df.iterrows():
    shap_contrib = np.array([float(x) for x in row['shap_values'].replace('\n','')[1:-1].split()])
    shap_contrib = shap_contrib[0: row['total_atoms_solute']]
    shap_contrib = shap_contrib + (-7.470815 / row['total_atoms_solute'])
    Shap_contrib.append(shap_contrib)
    pred_from_sum_shap.append(sum(shap_contrib))
df['Shap_contrib'] = Shap_contrib
df['pred_from_sum_shap'] = pred_from_sum_shap
```

```python
shap_max = []
shap_min = []
shap_std = []
for _, row in df.iterrows():
    #shap_max.append(np.max(row['Shap_contrib']))
    #shap_min.append(np.min(row['Shap_contrib']))
    #shap_std.append(np.std(row['Shap_contrib']))

    # Maybe also consider reciprocal?
    shap_max.append(np.max(row['Shap_contrib']) / np.abs(row['predicted']))
    shap_min.append(np.min(row['Shap_contrib']) / np.abs(row['predicted']))
    shap_std.append((np.std(row['Shap_contrib'])) / np.abs(row['predicted']))

    #shap_max.append(np.max(row['Shap_contrib']) * row['predicted'])
    #shap_min.append(np.min(row['Shap_contrib']) * row['predicted'])
    #shap_std.append(np.std(row['Shap_contrib']) * row['predicted'])
```

```python
df['shap_max'] = shap_max
df['shap_min'] = shap_min
df['shap_std'] = shap_std
```

```python
df['rxn_number'] = df['label'].apply(lambda x:int(x.split("_")[0][3:]))
```

```python
df2 = pd.read_csv('rxns_of_interest.csv')
df2.head(n=3)
```

```python
for rxn_number, subdf in df.groupby('rxn_number'):
    for solvent, subsubdf in subdf.groupby('can_smiles_solvent'):
        R = subsubdf[subsubdf.label.str.contains('_r')]
        P = subsubdf[subsubdf.label.str.contains('_p')]

        #R_shap_max, R_shap_min, R_shap_std = max(R['shap_max']), min(R['shap_min']), max(R['shap_std'])
        #P_shap_max, P_shap_min, P_shap_std = max(P['shap_max']), min(P['shap_min']), max(P['shap_std'])

        DGsolv_R, DGsolv_P = sum(R['predicted']), sum(P['predicted'])
        
        R_shap_max, R_shap_min, R_shap_std = sum(R['shap_max']), sum(R['shap_min']), sum(R['shap_std'])
        P_shap_max, P_shap_min, P_shap_std = sum(P['shap_max']), sum(P['shap_min']), sum(P['shap_std'])


        df2.loc[df2['solvent_smiles'] == solvent, 'DGsolv_R'] = DGsolv_R
        df2.loc[df2['solvent_smiles'] == solvent, 'DGsolv_P'] = DGsolv_P
        df2.loc[df2['solvent_smiles'] == solvent, 'DGsolv_PR'] = DGsolv_P - DGsolv_R
        
        df2.loc[df2['solvent_smiles'] == solvent, 'R_max'] = R_shap_max
        df2.loc[df2['solvent_smiles'] == solvent, 'R_min']= R_shap_min
        df2.loc[df2['solvent_smiles'] == solvent, 'R_std'] = R_shap_std
        
        df2.loc[df2['solvent_smiles'] == solvent, 'P_max'] = P_shap_max
        df2.loc[df2['solvent_smiles'] == solvent, 'P_min'] = P_shap_min
        df2.loc[df2['solvent_smiles'] == solvent, 'P_std'] = P_shap_std   

        df2.loc[df2['solvent_smiles'] == solvent, 'P-R_max'] = P_shap_max - R_shap_max
        df2.loc[df2['solvent_smiles'] == solvent, 'P-R_min'] = P_shap_min - R_shap_min
        df2.loc[df2['solvent_smiles'] == solvent, 'P-R_std'] = P_shap_std - R_shap_std

        df2.loc[df2['solvent_smiles'] == solvent, 'P+R_max'] = P_shap_max + R_shap_max
        df2.loc[df2['solvent_smiles'] == solvent, 'P+R_min'] = P_shap_min + R_shap_min
        df2.loc[df2['solvent_smiles'] == solvent, 'P+R_std'] = P_shap_std + R_shap_std

        df2.loc[df2['solvent_smiles'] == solvent, 'PR_max'] = P_shap_max * R_shap_max
        df2.loc[df2['solvent_smiles'] == solvent, 'PR_min'] = P_shap_min * R_shap_min
        df2.loc[df2['solvent_smiles'] == solvent, 'PR_std'] = P_shap_std * R_shap_std
```

```python
df2['ln_krel'] = np.log(df2['krel (kliq_to_kref)'])
```

```python
corr_coeffs = []

for rxn_idx, sub_df in df2.groupby('reaction_index'):
    #print(rxn_idx)
    corr_coeffs.append( [rxn_idx] + list(np.corrcoef([sub_df['DGsolv_R'], sub_df['DGsolv_P'], sub_df['DGsolv_PR'],
                       sub_df['R_max'], sub_df['R_min'], sub_df['R_std'], 
                       sub_df['P_max'], sub_df['P_min'], sub_df['P_std'],
                       sub_df['P-R_max'], sub_df['P-R_min'], sub_df['P-R_std'],
                       sub_df['P+R_max'], sub_df['P+R_min'], sub_df['P+R_std'],
                       sub_df['PR_max'], sub_df['PR_min'], sub_df['PR_std'],
                       sub_df['ln_krel']  ])[-1]) )

corr_coeffs = pd.DataFrame(corr_coeffs, columns = ['rxn_idx', 'DGsolv_R','DGsolv_P', 'DGsolv_PR',
                                                               'R_max', 'R_min', 'R_std', 
                                                               'P_max', 'P_min', 'P_std', 
                                                               'P-R_max', 'P-R_min', 'P-R_std', 
                                                                'P+R_max', 'P+R_min', 'P+R_std', 
                                                               'PR_max', 'PR_min', 'PR_std', 
                                                               'ln_krel']   
                          )
corr_coeffs
```

```python
corr_coeffs.to_csv('corr_coeffs.csv', index=False)
```
